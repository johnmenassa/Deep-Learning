##Dataset Description

Load diabets dataset from sklearn
[diabet dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html)


**dataset statistics**

image.png

## Prepare Dataset

```python
from sklearn import datasets
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

Split data into train/test set according to ratio 0.8/0.2.

```python
# Split the data into training/testing sets
train_num = int(len(diabetes_X) * 0.8)
diabetes_X_train = diabetes_X[:train_num]
diabetes_X_test = diabetes_X[train_num:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:train_num]
diabetes_y_test = diabetes_y[train_num:]
```

```python
len(diabetes_X_test), len(diabetes_X_train)
```

## Model Training

Create linear regression object

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

Train the model using the training sets

```python
regr.fit(diabetes_X_train, diabetes_y_train)
```

Make predictions using the testing set

```python
diabetes_y_pred = regr.predict(diabetes_X_test)
```

The learned coefficients $W$ for linear regression ($WX=Y$).

```python
print("Coefficients: \n", regr.coef_)
```

## Evaluation

```python
# import evaluation metric
from sklearn.metrics import mean_squared_error, r2_score
```

The mean squared error

```python
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
```

 $R^2$(coefficient of determination) regression score function.

Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).

```python
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))
```

# Logistic Regression (binary)

Load breast_cancer dataset from sklearn
[breast_cancer](https://scikit-learn.org/stable/datasets/toy_dataset.html#breast-cancer-dataset) for binary classification

image.png

##Prepare Dataset

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
X, y = load_breast_cancer(return_X_y=True) # (data, target)
```

```python
# Split the data into training/testing sets
train_num = int(len(X) * 0.8)
X_train = X[:train_num]
X_test = X[train_num:]

# Split the targets into training/testing sets
y_train = y[:train_num]
y_test = y[train_num:]
```

```python
len(X_train), len(X_test)
```

## Model Training

random_state controls the random seed

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)  
```

## Evaluation On the Test set

```python
print('prediction accuracy', clf.score(X_test, y_test))
```

###Get predictions on the test set

```python
clf.predict(X_test)
```

###Get prediction probability

```python
print('predicted probability for the first 10 test samples')
print(clf.predict_proba(X_test[0: 10]))
print('ground truth', y_test[0: 10])
```

#Logistic Regression (multinomial)

## Dataset Description

Load Iris plants dataset from sklearn
[Iris plants dataset](https://scikit-learn.org/stable/datasets/toy_dataset.html#iris-plants-dataset)

image.png

## Prepare Dataset

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
X_iris, y_iris = load_iris(return_X_y=True) # (data, target)
```

```python
# Split the data into training/testing sets
train_num = int(len(X_iris) * 0.8)
X_iris_train = X_iris[:train_num]
X_iris_test = X_iris[train_num:]

# Split the targets into training/testing sets
y_iris_train = y_iris[:train_num]
y_iris_test = y_iris[train_num:]
```

```python
len(X_iris_train), len(X_iris_test)
```

```python
y_iris_train
```

## Model Training & Test (using cross-entropy loss to handle  multi-class setting)

```python
clf_iris = LogisticRegression(random_state=0, multi_class='multinomial').fit(X_iris_train, y_iris_train)
```

```python
print('accuracy on test data', clf_iris.score(X_iris_test, y_iris_test))
```

```python
print('model prediction on the first 10 test samples')
print(clf_iris.predict(X_iris_test))
print('predicted probability for the first 10 test samples')
print(clf_iris.predict_proba(X_iris_test))
print('ground truth', y_iris_test)
```

## Model Training & Test (using one-vs-rest (OvR) scheme to handle multi-class)

```python
clf_iris_ovr = LogisticRegression(random_state=0, multi_class='ovr').fit(X_iris_train, y_iris_train)
```

```python
print('accuracy on test data', clf_iris_ovr.score(X_iris_test, y_iris_test))
```

```python
print('in ovr setting, model prediction on the first 10 test samples')
print(clf_iris_ovr.predict(X_iris_test[0:10]))
print('in ovr setting, predicted probability for the first 10 test samples')
print(clf_iris_ovr.predict_proba(X_iris_test[0: 10]))
print('ground truth', y_iris_test[0: 10])
```
